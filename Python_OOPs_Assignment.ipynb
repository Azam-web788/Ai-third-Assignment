**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

```python
class Student:
    def __init__(self , name , roll_number):
        self.__name = name
        self.__roll_number = roll_number
    def getter_name(self):
        return self.__name
    def setter_name(self , name):
        self.__name = name
    def getter_rollnumber(self):
        return self.__roll_number
    def setter_rollnumber(self , roll_number):
        self.__roll_number = roll_number
    
```

```python
s1.getter_name()
```

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class BankAccount:
    def __init__(self , balance = 0):
        self.__balance = balance
    def deposit(self , amount):
        self.__balance += amount
        return f"you deposit Rs.{amount}"
    def withdraw(self , amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("transaction completed successfully")
        else:
            print("Insufficient balance")
            
```

```python
bank = BankAccount()
```

```python
bank.deposit(1000)
```

```python
bank.withdraw(200)
```

**Q7:** Multilevel Inheritance ka use karein. Class `Animal` -> Class `Mammal` -> Class `Dog`. Dog class ke object se parent classes ke features ko access karein.

```python
class Animal:
    def __init__(self):
        self.type = "Living Being"

    def eat(self):
        print("Animal is eating")


class Mammal(Animal):
    def __init__(self):
        super().__init__()
        self.has_hair = True

    def walk(self):
        print("Mammal is walking")


class Dog(Mammal):
    def __init__(self):
        super().__init__()
        self.breed = "German Shepherd"

    def bark(self):
        print("Dog is barking")


dog1 = Dog()

print(dog1.type) 
dog1.eat()
dog1.walk()
dog1.bark()
```

**Q9:** Child class ke constructor (`__init__`) ke andar parent class ke constructor ko call karne ke liye `super()` ka use kaise hota hai? Aik practical example likhein.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print("Person constructor call hua")


class Student(Person):
    def __init__(self, name, age, roll_number):
        # Parent class ka constructor call
        super().__init__(name, age)

        self.roll_number = roll_number
        print("Student constructor call hua")

    def show_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll Number:", self.roll_number)
```

```python
s1 = Student("Azam", 20, 101)
```

```python
s1.show_info()
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
class Shape:
    def __init__(self, color):
        self.color = color

    def show_color(self):
        print("Color:", self.color)


class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def area(self):
        print("Circle Area:", 3.14 * self.radius * self.radius)


class Square(Shape):
    def __init__(self, color, side):
        super().__init__(color)
        self.side = side

    def area(self):
        print("Square Area:", self.side * self.side)
```

```python
c1 = Circle("Red", 5)
```

```python
s1 = Square("Blue", 4)
```

```python
c1.show_color()
```

```python
s1.show_color()
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def show(self):
        print(f"Point({self.x}, {self.y})")

p1 = Point(2, 3)
p2 = Point(4, 5)

p3 = p1 + p2 

```

```python
p3.show()
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
from abc import ABC, abstractmethod


class Car(ABC):

    def start_engine(self):
        print("Engine started... Vroom Vroom!")

    @abstractmethod
    def drive(self):
        pass


class SportsCar(Car):
    def drive(self):
        print("SportsCar is driving very fast.")


class FamilyCar(Car):
    def drive(self):
        print("FamilyCar is driving safely")

s = SportsCar()
f = FamilyCar()

s.start_engine()
s.drive()

print("-----")

f.start_engine()
f.drive()
```

**Q19:** Abstraction aur Encapsulation ke darmiyan kya wazeh farq (main difference) hai? Dono ke concepts ko short statements mei define karein.

# Abstractiion

    - Focus: What to do

# Encapsulation

    - Focus: Data security

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC, abstractmethod


class Payment(ABC):

    def payment_info(self):
        print("Payment process start ho raha hai...")

    @abstractmethod
    def pay(self, amount):
        pass


class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card 💳")


class UPIPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using UPI 📱")

```

```python
p1 = CreditCardPayment()
p2 = UPIPayment()

p1.payment_info()
p1.pay(5000)

p2.payment_info()
p2.pay(2000)
```
